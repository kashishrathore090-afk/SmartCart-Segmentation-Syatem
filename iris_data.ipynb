```python
from sklearn.datasets import load_iris
import seaborn as sns
from sklearn.cluster import KMeans
import pandas as pd
```

```python
df = load_iris()
```

```python
X = df.data
y = df.target
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1])
```

```python
# scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
#PCA

from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca_data = pca.fit_transform(X_scaled)
```

```python
wcss = []

for k in range(1,11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit_transform(pca_data)

    wcss.append(kmeans.inertia_)
```

```python
sns.lineplot(x=range(1,11), y=wcss, marker="o")
```

```python
# K MEANS

kmeans = KMeans(n_clusters = 3)
labels = kmeans.fit_predict(pca_data)

sns.scatterplot(x=X_scaled[:,0], y=X_scaled[:,1], c=labels)
```

```python
sns.scatterplot(x=pca_data[:,0], y=pca_data[:,1], c=labels)
sns.scatterplot(x=kmeans.cluster_centers_[:,0], y=kmeans.cluster_centers_[:,1], marker="x",c="red")
```

```python
kmeans = KMeans(n_clusters = 3,random_state=10)
labels = kmeans.fit_predict(pca_data)

sns.scatterplot(x=pca_data[:,0], y=pca_data[:,1], c=labels)
sns.scatterplot(x=kmeans.cluster_centers_[:,0], y=kmeans.cluster_centers_[:,1], marker="x",c="red",s=100)
```
